```python
from demo_include import *

import ipywidgets as widgets
from IPython.display import display
import os
from IPython.display import Image


style = {'description_width': 'initial'}
title = widgets.HTML(value="<h1>RAELLA Simulator&nbsp;&nbsp;&nbsp;&nbsp;</h1>")
rows_slider = widgets.IntSlider(value=512, min=64, max=2048, step=64, description='Number of rows', style=style)
adc_resolution_slider = widgets.IntSlider(value=7, min=4, max=16, step=1, description='ADC resolution', style=style)
center_offset_checkbox = widgets.Checkbox(value=False, description='Center Offset', style=style)
adaptive_weight_slicing_checkbox = widgets.Checkbox(value=False, description='Adaptive Weight Slicing', style=style)
speculation_enabled_checkbox = widgets.Checkbox(value=False, description='Enable speculation', style=style)
clear_output_checkbox = widgets.Checkbox(value=False, description='Clear previous output', style=style)
device_type_dropdown = widgets.Dropdown(options=['SRAM', 'Protonic'], value='SRAM', description='Device Type', style=style)


def display_all():
    display(Image(filename='raella.png'))
    top_box = widgets.HBox([title, run_button, clear_output_checkbox], layout=widgets.Layout(align_items='center'))
    sliders_box = widgets.VBox([rows_slider, adc_resolution_slider, device_type_dropdown])
    checkboxes_box = widgets.VBox([center_offset_checkbox, adaptive_weight_slicing_checkbox, speculation_enabled_checkbox])
    hbox = widgets.HBox([sliders_box, checkboxes_box])
    display(Markdown("""
        \n The Titanium Law:
        $$\\frac{ADC\\ Energy}{DNN}=\\frac{Energy}{Convert} \\times \\frac{Converts}{MAC} \\times \\frac{MACs}{DNN} \\times \\frac{1}{Utilization}$$
    """))
    display(top_box, hbox)

def run(b):
    # Clear previous output
    run_raella(
        adc_bits=adc_resolution_slider.value, 
        n_rows=rows_slider.value, 
        n_cols=512,
        center_offset=center_offset_checkbox.value,
        adaptive_weight_slicing=adaptive_weight_slicing_checkbox.value,
        speculation=speculation_enabled_checkbox.value,
        device_type=device_type_dropdown.value
    )
    if clear_output_checkbox.value:
        clear_output()
        display_all()
    plt.show()

run_button = widgets.Button(description="Run")
run_button.on_click(run)
display_all()

# 8Fx6F
# Range: 0.1 uS to 1.6uS. 6b for 100s
# Read voltage: 0.1-0.5V
# Intrinsic capacitance: 3.54 fF/um^2

# PROTONIC PROTONIC PROTONIC
```
